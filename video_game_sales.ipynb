### Q2: Taking the previous question into account, when was the "golden year" of Nintendo? List the video games of that year.

#### NOTE: It is clear that sales are produced during many more years, so we will take "golden year" as the year whose video games released have shown better results in the long term.

We will need to use the groupby method to aggregate the rows for each year. We will also have to filter the results to only include the video games which have been published by Nintendo.

```python
df_Nintendo = df_vgs[df_vgs["Publisher"] == "Nintendo"].sort_values("Year")

ser_global_sales = df_Nintendo.groupby("Year").sum()["Global_Sales"]


mpl.rcParams["figure.figsize"] = (12,7)
sns.set_style("darkgrid")

plt.plot(ser_global_sales, 'D-g')
plt.title("Nintendo Global Sales between 1980 and 2016")
plt.xlabel("Year")
plt.ylabel("Global Sales (in millions)")
```

As you can see, 2006 was the best year. Let's see which video games were released in this year:

```python
df_Nintendo[df_Nintendo["Year"] == 2006].sort_values("Global_Sales", ascending = False)
```

According to the results, it is clear that "Wii Sport" was the main cause of our conclusion.

### Q3: According to our analysis, "Action" was the most popular genre between 1980 and 2016. Which platform has the highest   number of sales in this genre? 

We will need to filter the results to only include action video games. Then, will need to use the groupby method to aggregate the rows for each Platform.

```python
df_Action = df_vgs[df_vgs["Genre"] == "Action"]
df_Action_Platform = df_Action.groupby("Platform")["Global_Sales"].sum().sort_values(ascending = False)

sample = df_Action_Platform.head(10)

sns.set_style("white")

sns.barplot(sample.index, sample, palette = "mako" )
plt.title("Global Sales of Action Video Games per Plarform between 1980 and 2016")
plt.xlabel("Platform")
plt.ylabel("Global Sales (in millions)")

sample
```

As you can see, PS3 has the highest  number of sales in action video games, followed by PS2 and X360.

### Q4: How many video games have been sold exclusively in one region?

We will need to filter those rows whose "Global_Sales" are the same of only one region sales.

In this case, we will distinguish between Japan, Europe and North America:

```python
filt_JP = (df_vgs["JP_Sales"] == df_vgs["Global_Sales"])

res_JP = df_vgs[filt_JP].shape[0]

filt_EU = (df_vgs["EU_Sales"] == df_vgs["Global_Sales"])

res_EU = df_vgs[filt_EU].shape[0]

filt_NA = (df_vgs["NA_Sales"] == df_vgs["Global_Sales"])

res_NA = df_vgs[filt_NA].shape[0]

print("Japan: {}. Europe: {}. North America: {}.".format(res_JP, res_EU, res_NA))
```

Let's see the different proportions:

```python
total = df_vgs.shape[0]

prop_JP = res_JP/total
prop_EU = res_EU/total
prop_NA = res_NA/total

print("Japan: {:.2f}%. Europe: {:.2f}%. North America: {:.2f}%.".format(prop_JP*100, prop_EU*100, prop_NA*100))
```

### Q5: In which region has the saga "Pokémon" sold more video games? 

First of all, we will create a function which will return us a list of "Pokémon" video games (just names). Then, we will use this list to filter the rows which are "Pokémon" video games.

```python
def pok_video_games(list):
    result = []
    for word in list:
        if (word.find("Pokemon") + word.find("Pokémon")) != -2:
            result.append(word)
            
    
    return result

list = pok_video_games(df_vgs.index)

df_Pokémon = df_vgs[df_vgs.index.isin(list)]

north_america = df_Pokémon["NA_Sales"].sum()
europe = df_Pokémon["EU_Sales"].sum()
japan = df_Pokémon["JP_Sales"].sum()
other = df_Pokémon["Other_Sales"].sum()

index = ["North_America", "Europe", "Japan", "Other Regions"]
values = [north_america, europe, japan, other]

plt.pie(values, labels=index, autopct='%1.1f%%', startangle = 180);
plt.title('"Pokémon" Video Game Sales in different regions between 1980 and 2016')
```

As can be seen, North America has the highest  number of "Pokémon" video game sales.
